```python
#作業1：印菱形
```

```python
number = int(input("菱形要有幾層呢？"))
if(int(number)%2 == 0):
    print("請輸入基數")
else:
    s = '*'
    for i in range(1, number+1, 2):
        print((s*i).center(number))
    for i in reversed(range(1, number-1, 2)):
        print((s*i).center(number))
```

```python
#作業2：印灰階圖片（原本的圖片是彩色）
```

```python
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open('pug.jpg')
gray=img.convert('L')
plt.figure('pug')
plt.imshow(gray,cmap ='gray')
plt.axis('off')
plt.show()
```

```python
for i in range(1, 10):
        for j in range(1, i+1):
            print("%d*%d=%d  " % (j, i, i*j), end="")
        print()
```
