# HW11-3 爬蟲-台灣銀行匯率

```python
#方法2: pandas
import pandas
import sqlite3 #和mySQL不同，不需要在本地端起一個服務，只要一個檔案
from datetime import datetime
dfs = pandas.read_html("http://rate.bot.com.tw/xrt?Lang=zh-TW")
currency = dfs[0]
currency = currency.ix[:,[0,1,2,3,4]]
currency.columns = [u"幣別",u"現金匯率-本行買入",u"現金匯率-本行賣出",u"即期匯率-本行買入",u"即期匯率-本行賣出"]
currency[u"幣別"] = currency[u"幣別"].str.extract('\((\w+)\)') #正規表達法
currency["Date"] = datetime.now().strftime("%Y-%m-%d")
currency["Date"] = pandas.to_datetime(currency["Date"])

with sqlite3.connect("currency.sqlite") as db:   #存成sqlite
    currency.to_sql("currency", con = db, if_exists ="append")
```

```python
currency.to_excel("currency.xlsx") #變成excel檔
```
