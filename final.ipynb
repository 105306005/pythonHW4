# 爬Dcard 

使用pyquery爬Dcard，將標題和內容變成excel檔

不過需要注意Mac系統Excel的CSV檔，會有亂碼問題，可以參考這個網址：https://blog.surveycake.com/02-06-948b8857b0db

```python
import requests
import pyquery
import pandas as pd
from openpyxl import Workbook
from openpyxl.compat import range
from openpyxl.utils import get_column_letter
```

```python
news_list =[]
article_date = []
article_title= [] 
article_content= []
article_data_url = []

print("爬Dcard")

dcard_url = 'https://www.dcard.tw/f'
req = requests.get(dcard_url)
page = req.text
pq = pyquery.PyQuery(page)

a_data = pq('.PostList_entry_1rq5Lf').find('a')
print(a_data.attr('href'))

for data in a_data: 
    article_data_url= "https://www.dcard.tw"+ pq(data).attr('href')
    print(article_data_url)
    req2 = requests.get(article_data_url)
    page2 = req2.text
    pq2 = pyquery.PyQuery(page2)
    article_title = pq2('.Post_title_2O-1el').text() 
    print(article_title)
    article_content = pq2('.Post_content_NKEl9d').text() 
    article_content = article_content.split("\n")
    while '' in article_content:
        article_content.remove('')
    article_content = ','.join(article_content) 
    print(article_content) 
    article_date = pq2('.Post_date_2ipeYS').text() 
    print(article_date)      
    print("-----------------------------------------------------")
    news_dict = {'日期' : article_date, '標題' : article_title, '文章內容' : article_content, '網址' : article_data_url}
    news_list.append(news_dict)
    
#轉成dataframe 
df = pd.DataFrame(news_list, columns=['日期', '標題', '文章內容', '網址'])
#儲存成csv檔
df.to_csv('dcard.csv')
print("爬完囉")   
```
