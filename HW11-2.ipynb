```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import jieba
from PIL import Image
import numpy as np
```

```python
sdg = open("song.txt","r",encoding="UTF-8").read()
```

```python
wordlist_after_jieba=jieba.cut(sdg, cut_all=False)
wl_space_split=" ".join(wordlist_after_jieba)
print(wl_space_split)
```

```python
font="wt005.ttf"
```

```python
stopwords= set(STOPWORDS)
a=["我","你","他"]
new_stopwords = stopwords.union(a)
```

```python
my_wordcloud = WordCloud(font_path = font,stopwords = new_stopwords, background_color="black").generate(wl_space_split)
plt.imshow(my_wordcloud)
plt.axis("off")
plt.show()
```
